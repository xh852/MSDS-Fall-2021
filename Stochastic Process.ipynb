## (c)
We categorize 9 rooms into three types:
1. A room with 2 doors
2. A room with 3 doors
3. A room with 4 doors

The transition matrix now shows the probability that 🐿 moves from a room with 2 doors to a room with 2 doors, 3 doors, 4 doors, from a room with 3 doors to a room with 2 doors, 3 doors, 4 doors and from a room with 4 doors to a room with 2 doors, 3 doors, 4 doors. 

```python
P_symmetry = np.array(
                    [[1/3,1/2,  0],
                     [2/3,1/4,4/5],
                     [0,  1/4,1/5]])
```

```python
lamda_, v_ = np.linalg.eig(P_symmetry)
lamda_ = np.round(lamda_,0)
v_ = np.round(v_,2)
lamda_,v_
```

```python
mu_ = -v_[:,0]
mu_ = mu_/np.sum(mu_)
mu_
```
